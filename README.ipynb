## 1.2 Introduction

In the last few years a very controversial and popular topic of discussion is the drug problem in Europe. The glorification of drugs in popular culture(music and movies), as well as the growth of the Internet and social media, have all spiked the recent surge in drug consumption and availability in Europe. To tackle this problem, countries have implemented various measures, such as rehabilitation or drug prohibition by law. Some go with a liberal approach, some with a conservative approach, and some countries do not do much to handle this issue. Our goal with this research, is to analize the current drug trends in Europe, assess the effectiveness of the measures that countries have taken, and determine the factors that contribute to the drug problem. Throughout this project, we will put emphasis on the deaths caused by overdosing from drugs, and which factors contribute to the deaths as well.

## 4. Data analysis

### 4.1 Overdose deaths

On this graph we can see that the most overdose deaths are in Germany and the UK, which is no surprise as they are the largest populations in Europe.

The data seems to best fit the student or Beta estimates, so we can use this to calculate Z-scores for each countries(for deaths per 10 000 people), so that we can find the outliers(both in high and low deaths).

### 4.2  Alcohol deaths

The predominant trend is that males tend to overdose from alcohol alot more than females. However, the alcohol deaths remain relatively the same - fluctuating around 300.

### 4.3 Underage overdoses

There is a positive trend in underage overdoses( from the year 2012 to 2017 it is constantly falling).

### 4.4  Elderly overdoses

Elderly overdoses have decreased dramatically in the 21st century.

### 4.4 OST (Opioid substitution treatment)

If we look at the most common demographic groups, the average person that seeks OST is:
- 30 to 34 years old
- Has secondary education
- Consumes drugs daily
- Has stable accomodation(with his parents)
- Is unemployed
- Referred by a friend/family member.

### 4.5 Users and suppliers

We can see that there is a significant correlation between the amount of users and suppliers of ecstasy, heroin and cocaine, not so much in Cannabis.

### 4.6 Purity

Cocaine has a very variable purity, so we analyezed the correlation of purity vs overdoses - the result seemed counter intuitive -- the bigger the purity - the more overdoses.

### 4.7 Price vs users

Through all drugs except cocaine(in countries with high deaths) we can see that the price does not reduce the amount of users - this is because the price increases are not very significant - except for cocaine in users of high death countries. This gives us the impression that no matter what the price is - people in Europe will still take drugs.
