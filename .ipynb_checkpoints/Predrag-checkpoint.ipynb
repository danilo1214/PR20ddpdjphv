```python
import pandas
from collections import defaultdict
import numpy as np
import matplotlib.pyplot as plt
f = pandas.read_excel("./tables/drugDeaths/deaths15-19.xlsx")
mrtvi=defaultdict(int)
for godina in range(1995,2018):
    for i,j in f.iterrows():
        brojevi=np.sum(f[godina])
        mrtvi[godina]=brojevi
godine=[v for v in mrtvi.keys()]
broj=[v for v in mrtvi.values()]
god=np.array(godine)
br=np.array(broj)
plt.figure(figsize=(25,10))
plt.bar(god,br,align="center",width=0.5,alpha=0.5)
plt.xlabel("Leto",fontsize=20)
plt.ylabel("Stevilo mrtvih od droge",fontsize=20)
plt.title("Stevilo mrtvih od droge leta 1995-2017 starih 15-19 let",fontsize=25)
```
