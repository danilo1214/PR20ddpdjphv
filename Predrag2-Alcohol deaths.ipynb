# Alcohol deaths
According to the 2018 National Survey on Drug Use and Health (NSDUH), 86.3 percent of people ages 18 or older reported that they drank alcohol at some point in their lifetime; 70.0 percent reported that they drank in the past year; 55.3 percent reported that they drank in the past month.

The data above is presenting alcohol deaths between 2011-2016.

```python
import pandas
from collections import defaultdict
import matplotlib.pyplot as plt
ft = pandas.read_csv("./tables/alcoholDeaths/alcoholDeathsfix.txt",delimiter=",")
male=defaultdict(list)
female=defaultdict(list)
for i,j in ft.iterrows():
    if(j["sex"]=="F"):
        for k in range(2011,2017):
            female[k].append(j[str(k)])
    else:
        for k in range(2011,2017):
            male[k].append(j[str(k)])
god=[i for i in range(2011,2017)]
muski=[sum(male[i]) for i in male.keys()]
zenski=[sum(female[i]) for i in female.keys()]
fig = plt.figure()
fig = plt.figure(figsize=(10,5))
ax = fig.add_axes([0,0,1,1])

ax.bar(god, muski, color="slateblue", label="Male", align='center', edgecolor="black")   # #2edf87(blue color)
ax.bar(god , zenski, color="pink", label="Female", align='center', edgecolor="black")
plt.title("Male and female deaths caused by alcohol", fontsize=18)
plt.xlabel("Year", fontsize=15)
plt.ylabel("Number of deaths", fontsize=15)
plt.legend(loc="best",shadow=True)
```

```python
po_godini=defaultdict(list)
for k in range(2011,2017):
    for i,j in ft.iterrows():
        po_godini[k].append(j[str(k)])
godine=[g for g in po_godini.keys()]
smrti=[sum(po_godini[s]) for s in po_godini.keys()]
plt.figure(figsize=(10,5))
plt.plot(godine,smrti,linewidth=2.0,marker=".",markersize=9,markeredgecolor="black")
plt.xlabel("Year",fontsize=15)
plt.ylabel("Number of deaths",fontsize=15)
plt.title("Number of alcohol deaths throughout years",fontsize=20)
plt.grid()    
```
